```python
with open('data.txt', 'w') as file:
    content = file.write("Hello, Welcome to Data Engineering:\n")
```

```python
with open('data.txt', 'a') as file:
    content = file.write("Adding more content.\n")
```

```python
with open('data.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
with open('data.txt', 'r') as file:
    for line in file:
        print(line)
        print(line.strip())
```
