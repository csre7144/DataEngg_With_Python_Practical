Read Data From CSV into Python

```python
import csv
with open('data.csv', 'r') as f:
    csv_reader = csv.reader(f)
    print(csv_reader)
```

```python
import csv
with open('data.csv', 'r') as f:
    csv_reader = csv.reader(f)
    print(csv_reader)
    data = [row for row in csv_reader]
    print(data)
```

```python
import csv
with open('data.csv', 'r') as f:
    csv_reader = csv.DictReader(f)
    for row in csv_reader:
        print(row)
```
