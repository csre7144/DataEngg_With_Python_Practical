Processing Python List

```python
data = [1, 2, 3, 4, 5]

squared = [x**2 for x in data]
print(squared)
```

```python
data = [1, 2, 3, 4, 5]

squared = [x**2 for x in data]
print(squared)

evens = [x for x in data if x % 2 == 0]
print(evens)

evens = [x for x in data if x % 2 != 0]
print(evens)
```

```python
data = [1, 2, 3, 4, 5]

squared = [x**2 for x in data]
print(squared)

odd = []
for x in data:
    if x % 2 != 0:
        odd.append(x)
print(odd)

evens = [x for x in data if x % 2 == 0]
print(evens)

evens = [x for x in data if x % 2 != 0]
print(evens)

total = sum(data)
print(total)

average = total / len(data)
print(average)
```
