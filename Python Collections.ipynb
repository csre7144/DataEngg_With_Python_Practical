```python
from collections import namedtuple, defaultdict, Counter

Employee = namedtuple('Employee', ['name', 'age', 'department'])
e1 = Employee('Alice', 30, 'HR')
e2 = Employee('Bob', 25, 'IT')

print(e1)
print(e2)

d = defaultdict(list)
d['HR'].append('Alice')
d['IT'].append('Bob')
print(d)
```
