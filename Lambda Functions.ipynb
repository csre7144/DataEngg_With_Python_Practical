Lambda Functions

```python
square = lambda x: x*x
print(square(5))
```

```python
add = lambda x, y: x + y
print(add(3, 4))
```

```python
numbs = [1, 2, 3, 4]

square = list(map(lambda x: x**2, numbs))
print(square)
```

```python
numbs = [1, 2, 3, 4, 5]

even = list(filter(lambda x: x % 2 == 0, numbs))
odd = list(filter(lambda x: x % 2 != 0, numbs))

print(even)
print(odd)
```
